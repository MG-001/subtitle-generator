##### Transcription

#### Load Model

```python
model = WhisperModel("small", device="cpu")
print("Model loaded successfully")
```

#### Detect language

```python
segments, info = model.transcribe("sample audio.wav")

print("Detected language:", info.language)
```

### Content

```python
for seg in segments:
    start = f"{seg.start:.2f}s"
    end = f"{seg.end:.2f}s"
    text = seg.text.strip()
    print(f"[{start} â†’ {end}] {text}")
    print("\n")
```
